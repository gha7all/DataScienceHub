```python
import pandas as pd
import numpy as np
from warnings import filterwarnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib

filterwarnings(action='ignore')
plt.style.use('ggplot')
pal = sns.color_palette()
matplotlib.rc('axes.formatter', useoffset=False)
```

```python
train = pd.read_csv('dataset/train.csv')
test = pd.read_csv('dataset/test.csv')
```

```python
train_info = pd.DataFrame({'variance': train.var(
), 'mean': train.mean(), 'standard deviation': train.std()})
train_info
```

```python
plt.figure(figsize=(25, 10))

sample = train_info.sample(80)
sns.barplot(x=sample.index, y=sample['variance'], color=pal[0])
```

```python
plt.figure(figsize=(25, 10))

sns.barplot(x=sample.index, y=sample['mean'], color=pal[0])
```

```python
# target distribution

target_dist = train.groupby('target', as_index=False)['target'].count()
sns.barplot(x=['0', '1'], y=target_dist['target'])
```

```python
cols = train.corr().nlargest(20, 'target')['target'].index

plt.figure(figsize=(16, 12))
sns.heatmap(train[cols].corr(), cmap='Greys', annot=True)
```

```python
# to see whether adding some statistical features could help

train_copy = train.drop(['target'], axis=1)

train['num_of_zeros'] = (train_copy == 0).sum(axis=1)
train['num_of_nonzeros'] = (train_copy != 0).sum(axis=1)
train['sum'] = train_copy.sum(axis=1)
train['mean'] = train_copy.mean(axis=1)
train['std'] = train_copy.std(axis=1)
train['var'] = train_copy.var(axis=1)
train['max'] = train_copy.max(axis=1)
train['min'] = train_copy.min(axis=1)

# train['negetives'] = 0
# train['positives'] = 0
```
